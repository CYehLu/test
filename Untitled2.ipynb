```python
data = data.reshape(13, 63, 304, 628)   # (t, var+z, y, x)

u = data[:, 0:10, :, :].reshape(13, 10, 304, 628)
v = data[:, 10:20, :, :].reshape(13, 10, 304, 628)
w = data[:, 20:30, :, :].reshape(13, 10, 304, 628)
t2 = data[:, 30, :, :].reshape(13, 304, 628)
u10 = data[:, 31, :, :].reshape(13, 304, 628)
v10 = data[:, 32, :, :].reshape(13, 304, 628)
height = data[:, 33:43, :, :].reshape(13, 10, 304, 628)
tc = data[:, 43:53, :, :].reshape(13, 10, 304, 628)
theta = data[:, 53:63, :, :].reshape(13, 10, 304, 628)
```

```python
x = np.arange(54.85012, 54.85012+628*0.20270273, 0.20270273)
y = np.arange(-7.37227, -7.37227+304*0.20270273, 0.20270273)
X, Y = np.meshgrid(x, y)

fig, ax = plt.subplots(nrows=13, ncols=4, figsize=(18, 35))

for itime in range(13):
    cs = ax[itime, 0].contourf(X, Y, u[itime, 2, :, :])
    ax[itime, 0].set_title('U, time = {}'.format(itime))
    fig.colorbar(cs, ax=ax[itime,0])
    
    cs = ax[itime, 1].contourf(X, Y, v[itime, 2, :, :])
    ax[itime, 1].set_title('V, time = {}'.format(itime))
    fig.colorbar(cs, ax=ax[itime,1])

    cs = ax[itime, 2].contourf(X, Y, theta[itime, 2, :, :])
    ax[itime, 2].set_title('theta, time = {}'.format(itime))
    fig.colorbar(cs, ax=ax[itime,2])
    
    cs = ax[itime, 3].contourf(X, Y, height[itime, 2, :, :])
    ax[itime, 3].set_title('height, time = {}'.format(itime))
    fig.colorbar(cs, ax=ax[itime,3])
    
plt.tight_layout()
```

```python
from difference import d_dx, d_dy
```

```python
dy = 6378000 * 0.20270273 * 3.14 / 180
radlat = np.deg2rad(np.arange(-7.37227, -7.37227+304*0.20270273, 0.20270273))
dx = np.cos(radlat) * dy
```

<font size="6">
$\frac{d}{dt}(\frac{\partial\theta}{\partial y}) = 
\frac{\partial}{\partial y}(\frac{d\theta}{dt}) - \frac{\partial u}{\partial y} \frac{d\theta}{dx}
 - \frac{\partial v}{\partial y} \frac{d\theta}{dy} - \frac{\partial w}{\partial y} \frac{d\theta}{dz}$ </font>

```python
# d(theta_y) / dt
dtheta_dy = np.zeros_like(theta)

nlev = 10
ntime = 13
for itime in range(ntime):
    for ilev in range(nlev):
        dtheta_dy[itime,ilev,:,:] = d_dy(theta[itime,ilev,:,:], dy)


dtheta_dy_dt = np.zeros_like(theta)

dtheta_dy_dt[0,:,:,:] = np.nan
dtheta_dy_dt[-1,:,:,:] = np.nan
for itime in range(1, ntime-1):
    dtheta_dy_dt[itime,:,:,:] = (dtheta_dy[itime+1,:,:,:] - dtheta_dy[itime-1,:,:,:]) / 2
```

```python
# d(theta_t) / dy
dtheta_dt = np.zeros_like(theta)

dtheta_dt[0,:,:,:] = np.nan
dtheta_dt[-1,:,:,:] = np.nan

for itime in range(1, ntime-1):
    dtheta_dt[itime,:,:,:] = (theta[itime+1,:,:,:] - theta[itime-1,:,:,:]) / 2
    

dtheta_dt_dy = np.zeros_like(theta)
for itime in range(ntime):
    for ilev in range(nlev):
        dtheta_dt_dy[itime,ilev,:,:] = d_dy(dtheta_dt[itime,ilev,:,:], dy)
```

```python
# du/dy * d(theta)/dx

du_dy = np.zeros_like(u)
dtheta_dx = np.zeros_like(theta)

for itime in range(ntime):
    for ilev in range(nlev):
        dtheta_dx[itime,ilev,:,:] = d_dx(theta[itime,ilev,:,:], dx)
        du_dy[itime,ilev,:,:] = d_dy(u[itime,ilev,:,:], dy)
        
dudy_dthetadx = du_dy * dtheta_dx
```

```python
# dv/dy * d(theta)/dy

dv_dy = np.zeros_like(v)
dtheta_dy = np.zeros_like(theta)

for itime in range(ntime):
    for ilev in range(nlev):
        dtheta_dy[itime,ilev,:,:] = d_dy(theta[itime,ilev,:,:], dy)
        dv_dy[itime,ilev,:,:] = d_dy(v[itime,ilev,:,:], dy)
        
dvdy_dthetady = dv_dy * dtheta_dy
```

```python
# dw/dy * d(theta)/dz

dw_dy = np.zeros_like(w)
dtheta_dz = np.zeros_like(theta)

for itime in range(ntime):
    for ilev in range(nlev):
        dw_dy[itime,ilev,:,:] = d_dy(w[itime,ilev,:,:], dy)        
    for ilev in range(1, nlev-1):
        dtheta_dz[itime,ilev,:,:] = (theta[itime,ilev+1,:,:] - theta[itime,ilev-1,:,:]) / 2
        
dwdy_dthetadz = dw_dy * dtheta_dz
```
