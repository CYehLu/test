```python
import pandas as pd
import numpy as np
```

```python
strike = np.arange(7400, 9400, 100, dtype='int')
Tt = np.random.randint(25, 37, size=strike.size)
F01 = np.ones_like(strike) * 8895
F02 = np.ones_like(strike) * 8840

df = pd.DataFrame({'strike': strike,
                   'Tt': Tt,
                   'F01': F01,
                   'F02': F02},
                  columns=['strike', 'Tt', 'F01', 'F02'])
df
```

```python
isLess31 = df['Tt'] < 31

g = df.groupby(by=isLess31)
```

```python
g.groups
```

```python
subdf1 = g.get_group(False)
subdf1[(subdf1['strike'] <= subdf1['F01']*1.15) & (subdf1['strike'] >= subdf1['F01'] * 0.85)]
```

```python
subdf2 = g.get_group(True)
subdf2[(subdf2['strike'] <= subdf2['F02']*1.15) & (subdf2['strike'] >= subdf2['F02'] * 0.85)]
```
