```python
import pandas as pd
import numpy as np
```

```python
# make some virtual data and index
np.random.seed(0)

data = np.arange(5, 105, 5)
rn = np.random.randint(1, 5, data.size)
data = np.repeat(data, rn)

date_index = pd.DatetimeIndex(freq='D', start='20170202', periods=17)
date_index = pd.to_datetime(np.repeat(date_index, 3))
```

```python
# example df
df = pd.DataFrame(data, index=date_index, columns=['data'])
df = df.reset_index()
df.head()
```

```python
gdf = df.groupby(by=['index', 'data'])
```

```python
gdf_keys = gdf.groups.keys()
```

```python
for key in gdf_keys:
    subdf = gdf.get_group(key)
    # ...some code...
    print(subdf)
    print()
```
